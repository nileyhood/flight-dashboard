```python
import streamlit as st
import pandas as pd
import plotly.express as px

st.set_page_config(page_title="Flight Pattern Dashboard", layout="wide")

st.title("✈️ Flight Pattern Visualization Dashboard")

# 파일 업로드
uploaded_file = st.file_uploader("Upload your CSV file", type=["csv"])

if uploaded_file:
    df = pd.read_csv(uploaded_file)

    st.subheader("📊 Raw Data")
    st.dataframe(df)

    # 컬럼 선택
    selected_columns = ['SEQ_NBR', 'BASE', 'SEQ_PATTERN', 'FLIGHT_PATTERN']
    df = df[selected_columns]

    st.subheader("🔍 Filter Options")

    # 필터 UI
    base_filter = st.multiselect("Select BASE", df['BASE'].unique())
    seq_filter = st.multiselect("Select SEQ_PATTERN", df['SEQ_PATTERN'].unique())

    filtered_df = df.copy()

    if base_filter:
        filtered_df = filtered_df[filtered_df['BASE'].isin(base_filter)]

    if seq_filter:
        filtered_df = filtered_df[filtered_df['SEQ_PATTERN'].isin(seq_filter)]

    st.subheader("📋 Filtered Data")
    st.dataframe(filtered_df)

    # 시각화
    st.subheader("📈 Visualization")

    col1, col2 = st.columns(2)

    with col1:
        fig1 = px.histogram(filtered_df, x="BASE", title="BASE Distribution")
        st.plotly_chart(fig1, use_container_width=True)

    with col2:
        fig2 = px.histogram(filtered_df, x="SEQ_PATTERN", title="SEQ Pattern Distribution")
        st.plotly_chart(fig2, use_container_width=True)

    # 패턴 카운트
    st.subheader("📊 Pattern Counts")
    count_df = filtered_df.groupby("FLIGHT_PATTERN").size().reset_index(name="Count")
    fig3 = px.bar(count_df, x="FLIGHT_PATTERN", y="Count")
    st.plotly_chart(fig3, use_container_width=True)

else:
    st.info("👆 Upload a CSV file to get started.")
```
